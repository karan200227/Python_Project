# Gender Distribution

```python
plt.figure(figsize=(5,5))
plt.title("Gender Distribution")
ax=sns.countplot(df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

# From the above chart we have analysed that : the number of females is more than the number of males

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Relatinships Between Parents Education And Students Score")
plt.show()
```

# From the above chart we have concluded that the parents education have the good impact on their kids scores 

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(gb1,annot=True)
plt.title("Relatinships Between Parents Marital Status And Students Score")
plt.show()
```

# From the above chart we have concluded that the parents Marital status  have negligible  impact on their kids scores

# Distribution Of Ethenic Groups

```python
print(df["EthnicGroup"].unique())
```

```python
GroupA=df.loc[(df['EthnicGroup']=="group A")].count()
GroupB==df.loc[(df['EthnicGroup']=="group B")].count()
GroupC=df.loc[(df['EthnicGroup']=="group C")].count()
GroupD=df.loc[(df['EthnicGroup']=="group D")].count()
GroupE=df.loc[(df['EthnicGroup']=="group E")].count()

l=["group A","group B","group C","group D","group E"]
mlist=[GroupA['EthnicGroup'],GroupB['EthnicGroup'],GroupC['EthnicGroup'],GroupD['EthnicGroup'],GroupE['EthnicGroup']]
plt.pie(x=mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethenic Groups ")
plt.show()
```
